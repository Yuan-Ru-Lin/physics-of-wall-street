```python
import matplotlib as mpl
# mpl.use('TkAgg')
mpl.rcParams.update({
    'figure.figsize': (4, 4),
    'figure.dpi': 150,
    'font.sans-serif': ['Microsoft JhengHei'],
})

import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
def gaussian(x, mean, sigma):
    return np.exp(-0.5 * ((x-mean) / sigma)**2) / np.sqrt(2*np.pi) / sigma
```

```python
%matplotlib notebook

fig, ax = plt.subplots()

fig.suptitle("å·´æ¥”åˆ©è€¶çš„è‚¡åƒ¹æ¨¡åž‹â€”â€”å¸¸æ…‹åˆ†ä½ˆæ›²ç·š")
ax.get_yaxis().set_label_text("æ©ŸçŽ‡åˆ†ä½ˆå‡½æ•¸")
ax.get_yaxis().set_ticks([])
ax.get_xaxis().set_label_text("è‚¡åƒ¹")

x = np.linspace(0, 300, 1000)
line, = ax.plot(x, gaussian(x, mean=100, sigma=10))

def animate(i):
    line.set_ydata(gaussian(x, mean=100, sigma=10 + 1 * i))
    line.set_label(f"ç¬¬ {i} å¹´")
    return [line, plt.legend()]


ani = animation.FuncAnimation(fig=fig, func=animate,
                              frames=range(1, 21),
                              interval=200, blit=True,
#                               repeat=False,
                              repeat_delay=1000,
                             )

# ani.save("movie.gif", writer="imagemagick")
fig.show()
```
